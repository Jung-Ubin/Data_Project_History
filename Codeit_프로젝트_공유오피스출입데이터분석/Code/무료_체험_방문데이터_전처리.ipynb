### 2. `user_uuid` 3번 초과로 등장하는지 확인
- 3일체험 유저들의 일자별 방문기록 기준인지 확인

#### 체험일자별 전체 비율 확인

```python
# user_uuid별 등장 횟수 집계
user_counts = df_cleaned['user_uuid'].value_counts()

# 그룹핑
one_day     = (user_counts == 1).sum()   # 1번만 등장
two_day     = (user_counts == 2).sum()   # 2번 등장
three_day   = (user_counts == 3).sum()   # 3번 등장
over_three  = (user_counts > 3).sum()    # 3번 초과

# 전체 유저 수
total_users = len(user_counts)

# 결과 정리
result = pd.DataFrame({
    "유저 수": [one_day, two_day, three_day, over_three],
    "비율(%)": [
        one_day / total_users * 100,
        two_day / total_users * 100,
        three_day / total_users * 100,
        over_three / total_users * 100
    ]
}, index=["1일 체험", "2일 체험", "3일 체험", "3일 초과"])

print(result.round(2))
```

#### 전체 데이터 중 **3일 초과 유저들의 비율**

```python
# user_uuid별 등장 횟수
user_counts = df_cleaned['user_uuid'].value_counts()

# 3번 초과 등장하는 유저 id 목록
over3_ids = user_counts[user_counts > 3].index

# 해당 유저들의 row만 추출
over3_rows = df_cleaned[df_cleaned['user_uuid'].isin(over3_ids)]

# 집계
num_users = len(over3_ids)                      # 유저 수
num_rows = len(over3_rows)                      # row 수
ratio_rows = num_rows / len(df_cleaned) * 100   # 전체 대비 %

print(f"3번 초과 유저 수: {num_users}")
print(f"이들이 차지하는 row 수: {num_rows}")
print(f"전체 데이터에서 차지 비율: {ratio_rows:.2f}%")
```

- **3번 초과 유저를 이상치라 판단**
- `3일 체험 신청자`라는 기준이 있기 때문에
- n일차에 취소 후 다음 날을 사용했겠다 라는 가능성이 있었지만,
- `약 5%`정도의 데이터여서 삭제로 판단함

```python
# 3번 초과 유저 id 추출
user_counts = df_cleaned['user_uuid'].value_counts()
over3_ids = user_counts[user_counts > 3].index

# df_cleaned에서 제외
df_cleaned_2 = df_cleaned[~df_cleaned['user_uuid'].isin(over3_ids)].reset_index(drop=True)
deleted_rows = df_cleaned['user_uuid'].isin(over3_ids).sum()

print(f"삭제된 유저 수: {len(over3_ids)}")
print(f"삭제된 행 수: {deleted_rows}")
print(f"삭제 후 데이터 shape: {df_cleaned_2.shape}")
```

#### 이상치 삭제 후 전체데이터 비율 확인

```python
# user_uuid별 등장 횟수 집계
user_counts = df_cleaned_2['user_uuid'].value_counts()

# 그룹핑
one_day     = (user_counts == 1).sum()   # 1번만 등장
two_day     = (user_counts == 2).sum()   # 2번 등장
three_day   = (user_counts == 3).sum()   # 3번 등장
over_three  = (user_counts > 3).sum()    # 3번 초과

# 전체 유저 수
total_users = len(user_counts)

# 결과 정리
result = pd.DataFrame({
    "유저 수": [one_day, two_day, three_day, over_three],
    "비율(%)": [
        one_day / total_users * 100,
        two_day / total_users * 100,
        three_day / total_users * 100,
        over_three / total_users * 100
    ]
}, index=["1일 체험", "2일 체험", "3일 체험", "3일 초과"])

print(result.round(2))
```

### 3. `first_enter_time`, `last_leave_time` 계산 했을 때 `stay_time`이 안맞는 경우

#### 계산이 안 맞는 행의 수와 비율

```python
# 1) datetime 변환 (안 되어 있으면)
df_cleaned_2['first_enter_time'] = pd.to_datetime(df_cleaned_2['first_enter_time'], errors='coerce')
df_cleaned_2['last_leave_time'] = pd.to_datetime(df_cleaned_2['last_leave_time'], errors='coerce')

# 2) stay_time 문자열을 timedelta로 변환
df_cleaned_2['stay_time_td'] = pd.to_timedelta(df_cleaned_2['stay_time'], errors='coerce')

# 3) first ~ last 차이 구하기
df_cleaned_2['calculated_stay'] = df_cleaned_2['last_leave_time'] - df_cleaned_2['first_enter_time']

# 4) 불일치 여부 (허용 오차 ±1분)
mask = (df_cleaned_2['stay_time_td'].notna()) & (df_cleaned_2['calculated_stay'].notna()) & (
    (df_cleaned_2['stay_time_td'] - df_cleaned_2['calculated_stay']).abs() > pd.Timedelta(minutes=1)
)

# 5) 결과 확인
mismatch_count = mask.sum()
total_rows = len(df_cleaned_2)
mismatch_ratio = mismatch_count / total_rows * 100

print(f"불일치 행 수: {mismatch_count}")
print(f"전체 데이터 대비 비율: {mismatch_ratio:.2f}%")

# 필요시 불일치 행만 따로 보기
df_mismatch = df_cleaned_2[mask][['user_uuid','stay_time','stay_time_td','first_enter_time','last_leave_time','calculated_stay']]
```

#### `first_enter_time` / `last_leave_time` 재 계산 후 `stay_time` 채우기

```python
# 1) 복사본 생성
df_cleaned_3 = df_cleaned_2.copy()

# 2) stay_time 문자열 → timedelta 변환
df_cleaned_3['stay_time_td'] = pd.to_timedelta(df_cleaned_3['stay_time'], errors='coerce')

# 3) first ~ last 차이 계산
df_cleaned_3['calculated_stay'] = df_cleaned_3['last_leave_time'] - df_cleaned_3['first_enter_time']

# 4) 불일치 행 찾기 (±1분 이상 차이 나는 경우)
mask = (df_cleaned_3['stay_time_td'].notna()) & (df_cleaned_3['calculated_stay'].notna()) & (
    (df_cleaned_3['stay_time_td'] - df_cleaned_3['calculated_stay']).abs() > pd.Timedelta(minutes=1)
)

print(f"수정할 row 수: {mask.sum()}")

# 5) 불일치한 행의 stay_time을 재계산 값으로 교체
df_cleaned_3.loc[mask, 'stay_time'] = df_cleaned_3.loc[mask, 'calculated_stay'].astype(str)

# 6) 보조 컬럼 삭제
df_cleaned_3 = df_cleaned_3.drop(columns=['stay_time_td', 'calculated_stay'])
```

#### 처리한 `df_cleaned_3` 에서 불일치 항목 확인

```python
# 1) stay_time 문자열 → timedelta 변환
df_cleaned_3['stay_time_td'] = pd.to_timedelta(df_cleaned_3['stay_time'], errors='coerce')

# 2) first ~ last 차이 계산
df_cleaned_3['calculated_stay'] = df_cleaned_3['last_leave_time'] - df_cleaned_3['first_enter_time']

# 3) 불일치 여부 체크 (±1분 이상 차이 나는 경우)
mask = (df_cleaned_3['stay_time_td'].notna()) & (df_cleaned_3['calculated_stay'].notna()) & (
    (df_cleaned_3['stay_time_td'] - df_cleaned_3['calculated_stay']).abs() > pd.Timedelta(minutes=1)
)

# 4) 결과 출력
mismatch_count = mask.sum()
total_rows = len(df_cleaned_3)
mismatch_ratio = mismatch_count / total_rows * 100

print(f"불일치 행 수: {mismatch_count}")
print(f"전체 데이터 대비 비율: {mismatch_ratio:.2f}%")

# 5) 실제 불일치 데이터 확인 (원하면)
df_mismatch = df_cleaned_3[mask][['user_uuid','first_enter_time','last_leave_time','stay_time']]
```

```python
# stay_time을 timedelta로 변환 후 문자열로 변환 (마이크로초 포함)
df_cleaned_3['stay_time'] = pd.to_timedelta(df_cleaned_3['stay_time'], errors='coerce') \
                                .astype(str) \
                                .str.split(' ').str[-1]  # "0 days" 제거

df_cleaned_3 = df_cleaned_3.drop(['stay_time_td', 'calculated_stay'], axis=1)
```

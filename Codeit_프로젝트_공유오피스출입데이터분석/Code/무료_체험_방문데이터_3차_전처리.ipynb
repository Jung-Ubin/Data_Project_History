```python
import pandas as pd

df = pd.read_csv(r"C:\Users\nuwba\Desktop\Codeit_4\중급2_프로젝트\중급_2\trial_visit_info.csv")

df
```

```python
df_clean = df.dropna()

df_clean
```

### 모든 행의 데이터가 같을 때 row 1개 남기고 나머지 삭제

```python
# 중복 제거 전 행 개수
before = df_clean.shape[0]

# 중복 제거 (모든 컬럼 동일 → 1개만 남기기)
df_unique = df_clean.drop_duplicates(keep='first')

# 중복 제거 후 행 개수
after = df_unique.shape[0]

# 삭제된 개수
deleted = before - after

print("중복 제거 전:", before)
print("중복 제거 후:", after)
print("삭제된 행 개수:", deleted)
```

```python
df2 = df_unique.drop(['stay_time', 'stay_time_second'], axis=1)

df2
```

### 데이터 정리

```python
# 각 유저별 동일날짜에 기록(row)이 2건 이상 있는 유저 수(+ 비율)
dup_counts = (
    df2.groupby(['user_uuid', 'date'])
       .size()  # row 개수
       .reset_index(name='row_count')
)

# 2건 이상인 경우 필터링
multi_record_users = dup_counts[dup_counts['row_count'] > 1]

# 유저 수 계산
n_multi_users = multi_record_users['user_uuid'].nunique()
total_users = df2['user_uuid'].nunique()
ratio = n_multi_users / total_users * 100

print(f"동일날짜 기록(row) 2건 이상 유저 수: {n_multi_users}")
print(f"전체 유저 수: {total_users}")
print(f"비율: {ratio:.2f}%")
```

```python
# 해당 유저들이 원본 데이터에서 차지하고 있는 데이터 비율
# 2건 이상 기록한 유저들의 user_id 집합
multi_user_ids = multi_record_users['user_uuid'].unique()

# 이 유저들이 차지하는 행(row) 수
multi_user_rows = df2[df2['user_uuid'].isin(multi_user_ids)].shape[0]

# 전체 행(row) 수
total_rows = df2.shape[0]

# 비율 계산
row_ratio = multi_user_rows / total_rows * 100

print(f"동일날짜에 2건 이상 기록한 유저가 차지하는 행 수: {multi_user_rows}")
print(f"전체 행 수: {total_rows}")
print(f"비율: {row_ratio:.2f}%")
```

```python
# 해당 유저들 타게팅으로 데이터 삭제
# 2건 이상 기록한 유저들의 user_id 집합
multi_user_ids = multi_record_users['user_uuid'].unique()

# 이 유저들의 행을 제외하고 새로운 데이터프레임 생성
df2_removed = df2[~df2['user_uuid'].isin(multi_user_ids)].reset_index(drop=True)

print(f"삭제 전 전체 행 수: {df2.shape[0]}")
print(f"삭제 후 전체 행 수: {df2_removed.shape[0]}")
print(f"삭제된 행 수: {df2.shape[0] - df2_removed.shape[0]}")
```

```python
# 동일 날짜 2건 이상 유저 정리확인
# 동일 날짜별 row 개수 다시 계산
dup_counts_removed = (
    df2_removed.groupby(['user_uuid', 'date'])
    .size()
    .reset_index(name='row_count')
)

# 2건 이상 있는 경우 필터링
multi_record_users_removed = dup_counts_removed[dup_counts_removed['row_count'] > 1]

if multi_record_users_removed.empty:
    print("동일 날짜에 2건 이상 이력이 있는 유저는 더 이상 존재하지 않습니다.")
else:
    n_multi_users = multi_record_users_removed['user_uuid'].nunique()
    print(f"동일 날짜에 2건 이상 기록한 유저 수: {n_multi_users}")
    display(multi_record_users_removed.head())
```

```python
# 문자열에서 공백 뒤(시간)만 추출
df2_removed['first_enter_time'] = df2_removed['first_enter_time'].astype(str).str.split().str[-1]
df2_removed['last_leave_time']  = df2_removed['last_leave_time'].astype(str).str.split().str[-1]

df2_removed
```

```python
def normalize_time_format(x):
    # NaN은 그대로 반환
    if pd.isna(x):
        return x
    
    # 문자열 변환
    x = str(x)
    
    # 23:59:59 처리 → .999999 붙이기
    if x == "23:59:59":
        return "23:59:59.999999"
    
    # 00:00:00 처리 → .000000 붙이기
    if x == "00:00:00":
        return "00:00:00.000000"
    
    # 마이크로초 없는 HH:MM:SS → .000000 붙이기
    if len(x) == 8:  # HH:MM:SS 형태
        return x + ".000000"
    
    # 이미 마이크로초 있는 값은 그대로
    return x

# 적용
df2_removed['first_enter_time'] = df2_removed['first_enter_time'].apply(normalize_time_format)
df2_removed['last_leave_time']  = df2_removed['last_leave_time'].apply(normalize_time_format)
df2_removed
```

```python
# 체류시간(초) 계산
df2_removed['stay_time_seconds'] = (
    pd.to_datetime(df2_removed['last_leave_time']) - pd.to_datetime(df2_removed['first_enter_time'])
).dt.total_seconds()

# 유저별 날짜순 정렬
df2_removed = df2_removed.sort_values(['user_uuid', 'date'])

# trial_day: 각 유저가 등장할 때마다 1씩 증가
df2_removed['trial_day'] = df2_removed.groupby('user_uuid').cumcount() + 1

# stay_time_second_total: 누적합
df2_removed['stay_time_second_total'] = df2_removed.groupby('user_uuid')['stay_time_seconds'].cumsum()

df2_removed
```

```python
# user_uuid와 trial_day 기준으로 정렬
df_sorted = df2_removed.sort_values(by=["user_uuid", "trial_day"], ascending=[True, True])

# 정렬된 결과 확인
df_sorted
```

```python
# 각 유저 별 trial_day의 1일차 때 first_enter_time과 trial_day 마지막 날 last_leave_time이 72시간 이 넘는 유저 수
#  유저별 첫 입실 시각과 마지막 퇴실 시각 구하기
user_time_range = df2_removed.groupby("user_uuid").agg(
    first_enter=("first_enter_time", "min"),
    last_leave=("last_leave_time", "max")
).reset_index()

# datetime 형변환
user_time_range["first_enter"] = pd.to_datetime(user_time_range["first_enter"])
user_time_range["last_leave"] = pd.to_datetime(user_time_range["last_leave"])

# 체류 기간 계산 (단위: 시간)
user_time_range["duration_hours"] = (user_time_range["last_leave"] - user_time_range["first_enter"]).dt.total_seconds() / 3600

# 72시간 이상인 유저
over_72_users = user_time_range[user_time_range["duration_hours"] >= 72]

# 결과 출력 (수 + 비율)
total_users = user_time_range["user_uuid"].nunique()
over_72_count = over_72_users["user_uuid"].nunique()
over_72_ratio = over_72_count / total_users * 100

print(f"72시간 이상 이용한 유저 수: {over_72_count}")
print(f"전체 유저 수: {total_users}")
print(f"비율: {over_72_ratio:.2f}%")
```

```python
# 각 유저별 trial_day 시작날 기준 72시간 오버된 유저 수(+ 비율)

# 유저별 첫 날짜와 마지막 날짜 구하기
user_date_range = df2_removed.groupby("user_uuid").agg(
    first_date=("date", "min"),
    last_date=("date", "max")
).reset_index()

# datetime 변환
user_date_range["first_date"] = pd.to_datetime(user_date_range["first_date"])
user_date_range["last_date"] = pd.to_datetime(user_date_range["last_date"])

# 기간 계산 (단위: 일)
user_date_range["duration_days"] = (user_date_range["last_date"] - user_date_range["first_date"]).dt.days

# 72시간(3일) 이상인 유저
over_72_users = user_date_range[user_date_range["duration_days"] >= 3]

# 전체 유저 수, 조건 만족 유저 수, 비율
total_users = user_date_range["user_uuid"].nunique()
over_72_count = over_72_users["user_uuid"].nunique()
over_72_ratio = over_72_count / total_users * 100

print(f"72시간(3일) 이상 이용한 유저 수: {over_72_count}")
print(f"전체 유저 수: {total_users}")
print(f"비율: {over_72_ratio:.2f}%")
```

```python
# 72시간(3일) 이상 이용한 유저 ID 리스트
over_72_user_ids = over_72_users["user_uuid"].unique()

# 해당 유저들 제거
df_removed_cleaned = df2_removed[~df2_removed["user_uuid"].isin(over_72_user_ids)].copy()

print("삭제 전 행 수:", len(df2_removed))
print("삭제 후 행 수:", len(df_removed_cleaned))
print("삭제된 유저 수:", len(over_72_user_ids))
```

```python
# 각 유저의 trial_day 가 가장 큰 값을 기준으로 유저별로 한 줄만 생성한 df 생성
# 기존 df_removed_cleaned여기서 유저별로 가져올 데이터는 ['site_id', 'date', 'stay_time_second_total', 'trial_day']

# 각 유저별 trial_day가 가장 큰 값 찾기
last_trial_df = (
    df_removed_cleaned
    .sort_values(["user_uuid", "trial_day"])  # 정렬
    .groupby("user_uuid")                     # 유저별 그룹
    .tail(1)                                  # 각 그룹에서 마지막 행만
)

# 필요한 컬럼만 추출
last_trial_df = last_trial_df[['user_uuid', 'site_id', 'date', 'stay_time_second_total', 'trial_day']]

# 확인
print("생성된 행 수:", len(last_trial_df))
```
